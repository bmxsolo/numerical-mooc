# __ Question 1 __

Nous avons les équations de Maxwell suivantes: 

(1) $\vec{\nabla} \times \vec{E}=-\frac{\partial \vec{B}}{\partial t}$   

(2) $\vec{\nabla} \times \vec{B}=\mu \vec{J}$                            

(3) $\vec{\nabla}.\vec{B}=0$    

Remarque : dans notre problème, on aura $\mu=\mu_{0}$.

De plus, $\vec{J}=\sigma \vec{E}$, et donc $\vec{E}=\frac{\vec{J}}{\sigma} $. On peut remplacer cette expression de $\vec{E}$ dans l'équation (1): 

$-\frac{\partial \vec{B}}{\partial t}=\vec{\nabla} \times \vec{E}=\frac{1}{\sigma}(\vec{\nabla} \times \vec{J}) $

Si on remplace $\vec{J}$ par la relation (2), on trouve:

$\vec{\nabla} \times \vec{E}=\frac{1}{\sigma \mu}[\vec{\nabla} \times (\vec{\nabla} \times \vec{B})]$ 

Cependant: $$\vec{\nabla} \times (\vec{\nabla} \times \vec{B})=\vec{\nabla}(\vec{\nabla}.\vec{B})-\Delta \vec{B}$$
(Voir cours d'électromagnétisme de bac 2)

Mais $\vec{\nabla}.\vec{B}=0$ par (3), on en déduit donc l'équation suivante:
$$ \boxed{\frac{\partial \vec{B}}{\partial t}=\frac{1}{\sigma\mu}\Delta \vec{B}}$$

Le champ magnétique obéit donc bien à une équation de diffusion, avec $\alpha=\frac{1}{\sigma\mu}$ qui est la diffusivité. 

# __Question 2 : Conditions aux bords__ 

$\bullet$ __1)__ : Premièrement, nous allons nous intéresser aux conditions au bord sur le bord externe du conducteur.

Tout d'abord, l'expression générale du champ magnétique $\vec{B}$ est la suivante :

$$\vec{B} = (B_{x}(x,y,z),B_{y}(x,y,z),B_{z}(x,y,z))$$ où $B_{x},B_{y},B_{z}$ sont les composantes du champ $\vec{B}$.

Dans ce problème, on peut assimiler notre conducteur homogène et infini dans la direction z à un solénoïde de longueur infini : en effet, les courants qui parcourent le conducteur sont des courants de surface qui peuvent être considérés comme des boucles de courant autour de ce conducteur. De plus, les courants de déplacement dans le conducteur sont négligés $\rightarrow$ On a vu, dans notre cours d'Électromagnétisme, que le champ magnétique à l'intérieur d'un solénoïde de longueur infini est uniquement dirigé selon l'axe z. Et on peut voir, vu l'orientation des axes x,y,z du problème, que le champ $\vec{B}$ sera orienté dans la direction inverse de l'axe z (c.f. règle de la main droite). Le champ magnétique n'aura donc pas de composante selon l'axe x et y :

$$\vec{B}=(0,0,-B_{z}(x,y,z))$$

Par symétrie (infini en z) et par translation selon l'axe z du problème, la composante $B_{z}$ ne dépendra pas de la variable z. Donc :

$$\vec{B}=(0,0,-B_{z}(x,y))$$


Pour trouver les conditions au bord, nous partons de l'équation de Maxwell-Ampère : 

$$\vec{\nabla} \times \vec{B} = \mu_{0}\vec{J}$$
Intégrons sur une surface (on va considérer cette surface comme une boite d'épaisseur infinitésimale qui se trouve à l'interface du conducteur et de l'extérieur): 

$$ \iint_S \vec{\nabla} \times \vec{B}. \vec{dS} = \mu_{0} \iint_S \vec{J}.\vec{dS}$$

Par le théorème du rotationnel, on peut se ramener à une intégrale sur le bord de la surface pour la partie gauche. Pour la partie droite, on peut se ramener à une intégrale de ligne le long du segment de la boite qui se trouve à l'interface entre les 2 milieux en introduisant la densité de courant $\vec{j}$ dont les unités sont $[j] = 1A/m$ : 

$$\oint_{\partial S} \vec{B}.\vec{dl} = \mu_{0}\int \vec{j}.\vec{dl}$$

À l'interface entre 2 milieux, par exemple avec le milieu 1 à l'intérieur du conducteur et milieu 2 à l'extérieur du conducteur, on aura l'équation suivante : 

$$\boxed{\vec{n_{21}} \times (\vec{B_2} - \vec{B_1}) = \mu_{0}\vec{j}}$$
où $\vec{n_{21}}$ est un vecteur unitaire dirigé du milieu 1 vers le milieu 2.

__Remarque__ : Pour notre problème ici présent, le $\vec{j}$ et le $\vec{J}$ sont équivalents : En effet, le courant de surface $\vec{J}$ est par définition  le nombre d'Ampères passant par mètre carré de section. Or ici, si on prend une section (aire) perpendiculaire à $\vec{J}$, on voit que considérer un nombre d'Ampère par unité de longueur revient au même car $\vec{J}$ ne forme qu'une "ligne" de courant à travers la surface. On a donc : 
$$\vec{n_{21}} \times (\vec{B_2} - \vec{B_1}) = \mu_{0}$$ 
car $j = J =1$. 

$\bullet$ Prenons par exemple $\vec{n_{21}}=\vec{u_y}$ et donc $\vec{j}=j_x\vec{u_x}$ (bord du haut du conducteur). Sachant que le champ $\vec{B}$ est nul à l'extérieur ($\vec{B_1}=\vec{0}$), on trouve : 

$$ -(\vec{u_y} \times -B_{z}\vec{u_z})= \mu_{0}\vec{u_x}$$

$$B_z\vec{u_x}= \mu_{0}\vec{u_x}$$
et donc : 

$$B_z=\mu_{0}$$ au bord extérieur. On peut appliquer ce raisonnement pour chaque côtés, et on obtiendra le résultat suivant : 

$$\boxed{\vec{B}= -\mu_{0}(0,0,1)}$$ aux bords extérieurs du conducteur.



$\bullet$ __2)__ : Pour la condition au bord sur le bord du supraconducteur, nous allons utiliser le fait que la composante tangentielle du champ éléctrique est continue au travers de la surface. 

La condition de continuité sur la composante tangentielle du champ électrique s'écrit de la manière suivante: 

$$\vec{n}_{21}\times (\vec{E_2}-\vec{E_1})=\vec{0}$$

où l'indice "1" est associé au milieu du supraconducteur, et l'indice "2" est quant à lui associé à la partie du conducteur, le vecteur $\vec{n}_{21}$ est le vecteur unitaire qui va du milieu 1 vers le milieu 2 tout en étant orhtogonal à la frontière se trouvant entre les deux milieux, $\vec{E_1}$ est le champ électrique dans le supra conducteur et $\vec{E_2}$ est le champ électrique à l'intérieur du conducteur.

Cependant, on sait que $\sigma_1=\infty$ à l'intérieur du supraconducteur, et que $\vec{E_1}=\frac{\vec{J_1}}{\sigma_1}$ (où $\vec{J_1}$ est le courant surface dans le supraconducteur), donc on en déduit que le champ électrique à l'intérieur du supraconducteur est nul, $\vec{E}_{1}=0$.

L'équation de continuité devient: $$\vec{n}_{21}\times\vec{E_2}=\vec{0}$$

Or le champ électrique dans le conducteur est donné par $\vec{E_2}=\frac{\vec{J_2}}{\sigma_2}$ (où $\vec{J_2}$ est le courant de surface dans le conducteur au voisinage de la paroi ), on a donc: 

$$\vec{n}_{21}\times\frac{\vec{J_2}}{\sigma_2}=\vec{0}$$

Qu'est-ce que cette relation implique pour notre problème (*Remarque* : Par soucis de notation , nous ne noterons plus l'indice $"2"$)? Regardons d'abord le côté supérieur du supraconducteur : le vecteur unitaire dans ce cas là est équivalent à $\vec{u_{y}}$. Après avoir effectué le rotationnel, on obtient le système d'équation suivant : 

$$\vec{u_{x}}. \frac{J_{z}}{\sigma} = 0 .\vec{u_{x}}$$

$$-\vec{u_{y}}. 0 = 0 .\vec{u_{y}}$$

$$-\vec{u_{z}}. \frac{J_{x}}{\sigma} =0 .\vec{u_{z}}$$

Et en développant $\vec{\nabla} \times \vec{B} = \mu_{0} \vec{J}$ on obtient : 

$$\partial_{y}B_{z}-\partial_{z}B_{y}=0$$  

$$\partial_{z}B_{x}-\partial_{x}B_{z}=\mu_{0} J_{y}$$  avec $B_{x}=B_{y}=0$ et $J_{x}=0$.

Etant donné que l'on regarde les équations sur la partie supérieure, on va s'intéresser à la dépendance de $\vec{B}$ selon l'axe y au voisinage du bord. On obtient donc comme condition : 

$$\boxed{\partial_{y}B_{z}=0}$$.

On peut raisonner de la même manière pour le côté inférieur du supraconducteur, et on obtiendra donc la même condition. 

Pour les côtés droits et gauches, nous pouvons aussi raisonner de la même manière mais en nous intéressant à la dépendance de $\vec{B}$ selon l'axe x au voisinage du bord. On obtiendra pour ces côtés la condition suivante : 

$$\boxed{\partial_{x}B_{z}=0}$$

$\bullet$ Champ magnétique à l'intérieur du supraconduteur : 

On sait que dans un supraconducteur, le champ magnétique est nul car il y a phénomène d'exclusion totale de $\vec{B}$ (Voir *Effet Meissner : https://fr.wikipedia.org/wiki/Effet_Meissner*)
Dans notre code, on devra donc imposer une condition supplémentaire : le champ $\vec{B}$ doit être nul à l'intérieur du supraconducteur à tout temps $t$. 

# __ Question 3 __

Pour calculer l'évolution du champ magnétique dans le conducteur, nous allons utiliser la méthode d'Euler (forward) dans le temps et avec une discrétisation centrée d'ordre 2 pour le laplacien, comme demandé dans l'énoncé. Pour cela, nous partons de l'équation de diffusion suivante : 

$$\frac{\partial \vec{B}}{\partial t}=\frac{1}{\sigma\mu_{0}}\Delta \vec{B}$$ que nous discrétisons de la manière suivante : 

$$\frac{B^{n+1}_{i,j} - B^n_{i,j}}{\Delta t} = \frac{1}{\sigma \mu_{0}} (\frac{B^n_{i+1, j} - 2B^n_{i,j} + B^n_{i-1,j}}{\Delta x^2} + \frac{B^n_{i, j+1} - 2B^n_{i,j} + B^n_{i,j-1}}{\Delta y^2})$$

*Remarque* : $x$ correspond à l'indice $i$, $y$ correspond à l'indice $j$ et les $n$ correspondents aux étapes de temps. Pour stocker nos valeurs, nous utiliserons la convention selon laquelle nous stockons les coordonnées $y$ dans les colonnes du tableau, et nous stockons les coordonées $x$ selon les lignes du tableau $\rightarrow$ tout élément $B_{i,j}$ sera stocké dans le format $B[j,i]$. 

Connaissant la valeur du champ $\vec{B}$ en tout point pour un $n$, on peut connaitre la valeur du champ en $n+1$ : 

$$B^{n+1}_{i,j}= B^n_{i,j} + \frac{\Delta t}{\sigma\mu_{0}} (\frac{B^n_{i+1, j} - 2B^n_{i,j} + B^n_{i-1,j}}{\Delta x^2} + \frac{B^n_{i, j+1} - 2B^n_{i,j} + B^n_{i,j-1}}{\Delta y^2})$$

Pour implémenter les conditions de Neumann au bord du supraconducteur, on utilise le fait que $\partial_{y}B_{z}=\partial_{x}B_{z}=0$ (les dérivées s'annulent). Et donc les valeurs de $B$ juste atour de la surface du supraconducteur seront égales aux valeurs de $B$ à une discrétisation plus loin autour cette surface. Nous condidérons que $\vec{B}=\vec{0}$ sur la surface du supraconducteur car la surface fait partie intégrante de celui-ci (c'est un choix) et que les conditions de Neumann sont valables au voisinage de cette surface.

```python
#On importe les fonctions et les fonctionnalités de Python qui vont nous servir pour notre code

import numpy
from matplotlib import pyplot
%matplotlib inline
from matplotlib import rcParams, cm
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

Comme dit précédemment, étant donné que l'on a considéré les conditions de Neumann au voisinage du supraconducteur, on va considérer que le champ magnétique $\vec{B}$ est nul à l'intérieur du supraconducteur et nul au bord de celui-ci, et nous implémenterons les conditions à partir d'une discrétisation un pas plus loin que le bord du supraconducteur. 

```python
#Nous implémetons notre fonction pour la diffusion du champ B.
def diff(B, nt, alpha, dt, dx, dy):
    
    for n in range(nt):                                        # attention : B[j,i]
        Bn = B.copy()
        
        #Conditions aux bords externes du conducteur :
        B[0,:] = -mu0
        B[-1,:]= -mu0
        B[:,0] = -mu0
        B[:,-1] = -mu0
        
        B[1:-1,1:-1] = Bn[1:-1,1:-1] + alpha *\
            (dt/dy**2 * (Bn[2:,1:-1] - 2*Bn[1:-1,1:-1] + Bn[:-2,1:-1]) +\
             dt/dx**2 * (Bn[1:-1,2:] - 2*Bn[1:-1,1:-1] + Bn[1:-1,:-2]))
         
        
        #par les données du problème, on sait que le supraconducteur se trouve dans [40:81,40:81]
        #Condition : B=0 à l'intérieur et au bord du supraconducteur   : 
        
        B[int((ny-1)*1/3):int((ny-1)*2/3)+1,int((nx-1)*1/3):int((nx-1)*2/3)+1] = 0  #Condition : B=0 à l'intérieur et au bord du supraconducteur   
                                                                                    
        #Condition de Neumann aux bords du supraconducteur :
        
        #bord gauche
        B[int((ny-1)*1/3)-2:int((ny-1)*2/3)+3,int((nx-1)*1/3)-1]=B[int((ny-1)*1/3)-2:int((ny-1)*2/3)+3,int((nx-1)*1/3)-2]
        #bord droit
        B[int((ny-1)*1/3)-2:int((ny-1)*2/3)+3,int((nx-1)*2/3)+1]=B[int((ny-1)*1/3)-2:int((ny-1)*2/3)+3,int((nx-1)*2/3)+2]
        #bord du bas
        B[int((ny-1)*1/3)-1,int((nx-1)*1/3)-2:int((nx-1)*2/3)+3]=B[int((ny-1)*1/3)-2,int((nx-1)*1/3)-2:int((nx-1)*2/3)+3]
        #bord du haut
        B[int((ny-1)*2/3)+1,int((nx-1)*1/3)-2:int((nx-1)*2/3)+3]=B[int((ny-1)*2/3)+2,int((nx-1)*1/3)-2:int((nx-1)*2/3)+3]
        
        #remarque : int(ny-1)*1/3 = 40 , int(ny-1)*2/3 = 80 et ny=nx=121
    return B
```

```python
L1 = 0.015       #largeur du conducteur en m(dans la direction x)
L2 = 0.015       #hauteur du conducteur en m(dans la direction y )

nx = 121         #nombre de point dans la direction x pour discrétiser les sections du domaine
ny = 121         #nombre de point dans la direction y pour discrétiser les sections du domaine
nt = 5000        #nombre de pas de temps 

dx = L1/(nx-1)   #longueur de discrétisation en x (en m)
dy = L2/(ny-1)   #longueur de discrétisation en y (en m)

x = numpy.linspace(0,L1,nx) 
y = numpy.linspace(0,L2,ny)

mu0 = 1.26*10**(-6)       #perméabilité du vide (en N/A^2)
sigma = 5.96*10**7        #conductivité (en S/m)
alpha = 1./(mu0*sigma)    #alpha est le coefficient de diffusion thermique pour l'équation de diffusion du champ B
```

$\bullet$ Dans l'énoncé, on nous demande de prendre un pas de temps $\Delta t = 2.5 \times 10^{-7}s$. Cependant il faut faire attention avec les pas de temps que l'on prend car nous pourrions avoir des problèmes de stabilité pour notre équation de diffusion. 

Nous avons vu dans le cours que, dans le cas 2D, l'équation de diffusion sera stable si : 

$$\alpha \frac{\Delta t}{(\Delta x)^2} + \alpha \frac{\Delta t}{(\Delta y)^2} < \frac{1}{2}$$

Dans notre problème, $\Delta x = \Delta y$. On peut donc écrire cette condition comme suit : 

$$ \alpha \frac{\Delta t}{(\Delta x) ^2} < \frac{1}{4}$$ 

$$\Delta t < \frac{1}{4} (\Delta x)^2 \sigma \mu_{0}$$

En entrant les valeurs numériques, on obtient : $\Delta t < 2.933 \times 10^{-7}$. 
Le $\Delta t$ donné est donc valable pour notre problème, il n'y aura pas de problème de stabilité.

Retournons à présent au calcul de diffusion : 

```python
Bi = numpy.zeros((ny,nx))    #initialement, le champ B est nul partout dans le conducteur
dt = 2.5*10**(-7)  
B = diff(Bi.copy(), nt, alpha, dt, dx, dy)
```

```python
#On créé un graphique qui permet d'avoir une visulalisation de la diffusion magnétique

pyplot.figure(figsize=(7,6))                           #taille du graphique
pyplot.title('Diffusion du champ magnétique dans le conducteur')
pyplot.contourf(x,y,B,100,cmap=cm.viridis)             #fonction coutourf qui permet de d'afficher la diffusion et qui prend 
pyplot.xlim(0, 1.5*10**-2);                            #comme paramètre x,y et B et on utilise le colormap Viridis
pyplot.ylim(0, 1.5*10**-2);
pyplot.xlabel(' largeur du conducteur (x en m)')
pyplot.ylabel('hauteur du contucteur (y en m)')
pyplot.colorbar();                                     
```

Remarquons qu'après un temps $nt \times dt$ le champ magnétique s'est diffusé à l'intérieur du conducteur sans jamais pénétrer à l'intérieur du supraconducteur, comme attendu. De plus, on obsèrve que le champ magnétique tend à s'uniformiser après un certain temps.

On nous demande la valeur du flux du champ B  à $t = 1.25 \times 10^{-2}s$ à travers la section $z=0$ du conducteur.

Par définition, le *flux du champ magnétique* à travers une surface est donné par : 

$$\Phi_{B}=\int_S \vec{B} . \vec{dS}$$

Cependant, ici notre problème est discrétisé, on va donc effectuer une somme à la place d'une intégrale : 

$$\Phi_{B} = \Sigma_{k} B_{k}. dS_{k}$$ où $B_{k}$ est le champ magnétique associé à l'élément de surface $dS_{k}$.  

*Remarque* :

$\bullet$ Nous prenons des vecteurs unitaires associés aux élements de surface qui sont parralèles à la direction du champ magnétique $\vec{B}$ dans le conducteur,  et donc $\cos(\theta)=1$ dans le produit scalaire $\vec{B} . \vec{dS}$.

$\bullet$ Dans notre configuration, les $B_{k}$ sont les $B_{i,j}$ et les $dS_{k}$ sont les $dS_{i,j}$ qui sont des éléments de surface pour lesquels les centres correspondent aux valeurs $B_{i,j}$.

Pour faire ce calcul de flux à travers le conducteur en $z=0$, nous allons séparer notre problème en 3 : 

1) une somme sur les coins (somme 1 en rouge)

2) une somme sur les bords du conducteur sans prendre en compte les coins (somme 2 en vert) 

3) une somme sur l'intérieur du conducteur sans compter les bords extérieurs (somme 3 en bleu)

Décrivons plus précisément la façon dont on va effectuer nos sommes : 

1) somme 1 : 

A chaque coin du conducteur, nous avons une valeur de $B_{i,j}$. On va "entourer" nos coins d'élément de surface $dx^2$ (car dx=dy dans notre problème)(petit cadre rouge sur notre schéma). Cependant, on remarque qu'une partie de la surface n'est pas comprise dans le conducteur, et on ne doit alors pas prendre en compte le flux (du $B_{i,j}$ associé à cet élément de surface) traversant cette surface qui est hors du conducteur (la surface que l'on va considérer est donc de $\frac{dx}{2}\times \frac{dx}{2}$). C'est pourquoi cette partie là est affichée en blanc sur le schéma. De plus le champ B est nul à l'extérieur du conducteur. La somme 1 est donc donnée par:

$$Somme\,1= (B[0,0] + B[0,-1] + B[-1,-1] + B[-1,0]) \times \frac{dx^2}{4}$$


2) somme 2:

En excluant les coins, nous regardons le flux associé aux bords externes du conducteur. Le long de ces bords, on a des valeurs de $B_{i,j}$. Comme pour la somme 1, on associe à ces valeurs de champs magnétique un élément de surface d'aire $dx^2$. Cependant, pour ce cas-ci, la moitié de cette surface associée se trouve à l'extérieure du conducteur. La surface que l'on va considérer pour chaque élément de champ magnétique est donc $dx\times\frac{dx}{2}$. A nouveau, nous avons affiché en blanc cette partie dont on ne tient pas compte. On verra plus loin dans le code comment est implémentée cette somme.

3) somme 3:

Notons d'abord que la partie en jaune du schéma correspond au supraconducteur.
Ici, nous ne tenons pas compte des bords externes du conducteur et nous allons cacluler le flux pour l'intérieur du conducteur. Comme avant, nous associons à chaque élément de champ magnétique un élément de surface d'aire $dx\times dx = dx^2$.

__REMARQUE __ Il est important de constater que pour les valeurs de champ magnétique au bord du supraconducteur (qui seront également associés à des éléments de surface), on va leur associer un flux magnétique nul (car B est nul sur ces bords). On sous-estime donc le flux magnétique total à travers la surface en $z=0$. En effet, on voit sur le schéma qu'il y aura une zone blanche de surface dont on ne tiendra pas compte du fait que $B=0$ sur le bord du supraconducteur et qu'à chaque $B$ correspond un élement $dx \times dx$.  

Ensuite, pour calculer le flux total de $\vec{B}$ à travers la surface, nous aurons qu'à sommer les 3 différents flux représentés par *somme 1, somme 2, somme 3*.

Implémentons cela à présent : 

```python
#dx=dy dans notre problème
somme1=(B3[0,0]+B3[0,-1]+B3[-1,-1]+B3[-1,0])*(dx**2/4)   #somme 1 est la somme sur les coins
```

```python
somme2=0                                #on initialise la somme 2 à 0 
for n in range (1,nx-1):                #on calcul la somme 2 sur les bords externes sans tenir compte des coins
    somme2 = somme2 + B3[0,n]*(dx**2/2)   #On regarde d'abord 1 seul bord (bord du bas)

somme2=4*somme2                         #comme le problème est complètement symétrique, on peut appliquer ce raisonnement pour
                                        #tous les bords (4)
```

```python
somme3=0                                #on initialise la somme 3 à 0
for n in range (1,nx-1):                #nous allons parcourir tous les éléments B[j,i] qui sont à l'intérieur du conducteur
    for m in range (1,nx-1):
        somme3 = somme3 + B3[n,m]*dx**2
```

```python
print('la somme 1 est égale à', somme1, 'Wb') #[flux]=Wb= Weber : unité du flux magnétique 
print('la somme 2 est égale à', somme2, 'Wb')
print('la somme 3 est égale à', somme3, 'Wb')
```

```python
Flux_total = somme1 + somme2 + somme3
print('Le flux total du champ magnétique (qui est probablement sous-évalué à cause de la somme 3) à travers le conducteur à          t = 1.25 10^(-2)s vaut',Flux_total,'Wb')
```

Cette petite valeur du flux magnétique peut se justifier par le fait que le courant de surface appliqué au bord externe est faible ($1A/m^{2}$) et on a une petite surface pour le conducteur ($ 1,5cm \times 1,5cm$).

De plus, on sait qu'il y aura uniformisation du champ magnétique après un certain temps $t$. On peut donc aisément calculer le flux maximal que l'on devrait atteindre après uniformisation. En effet, on sait qu'après uniformisation le champ magnétique vaudra $-\mu_{0}$ partout dans le conducteur, sauf dans le supraconducteur où il sera toujours nul. On peut donc calculer le flux maximal de la facon suivante : 

$$\Phi_{B_{max}}=-\mu_{0} \times Aire$$ où $Aire$ est l'aire du conducteur (sans le supraconducteur). Cette aire est donnée par : 

$$Aire = 0.015 \times 0.015 - 0.005 \times 0.005 = 0.0002 m^{2}$$

On obtient donc : 

$$\Phi_{B_{max}}=-\mu_{0} \times Aire = -2.52.10^{-10} Wb$$

Comme on aurait pu s'y attendre, au temps $t= 1.25.10^{-2}s$ nous sommes proches de l'uniformisation complète du champ magnétique dans le conducteur car notre valeur du flux magnétique à ce temps là ($-2.50405312482.10^{-10} Wb$) est proche de la valeur du flux magnétique maximal après uniformisation.
